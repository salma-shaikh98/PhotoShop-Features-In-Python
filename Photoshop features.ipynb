```python
from PIL import Image
```

```python
img=Image.open("Desktop/pingu.jpg")
```

```python
img.size
```

```python
img.save("E:/College/MSC-1/ASIP/proj/pingu.pdf")
```

```python
img.resize((250,250)) #hard resize
```

```python
img.thumbnail((200,200)) #soft resize
img
```

```python
from PIL import ImageEnhance
```

```python
enhancer=ImageEnhance.Sharpness(img) #sharpness of image
```

```python
enhancer.enhance(8) #1 for original, 0 from blur
```

```python
img.transpose(Image.FLIP_LEFT_RIGHT) #flipping image
```

```python
img.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
img.transpose(Image.ROTATE_270)
```

```python
#crop image

img.crop(box=(40,40,90,90))
```

```python
#increase color of an image

color=ImageEnhance.Color(img)
color.enhance(4)
```

```python
#Brightness

brightness=ImageEnhance.Brightness(img)
brightness.enhance(4)
```

```python
#contrast

contrast=ImageEnhance.Contrast(img)
contrast.enhance(2)
```

```python
#blur

from PIL import ImageFilter

img.filter(ImageFilter.GaussianBlur(radius=3))
```
